```python
import pandas as pd
import numpy as np
```

```python
sales_series = pd.Series([0, 5, 155, 0 ,518], index=['coffee', 'coffee', 'tea', 'coconut', 'sugar'], name='Sales')
sales_series
```

```python
#sales_series.loc[ LOGICAL TEST ]

sales_series.loc[sales_series > 0]
```

```python
mask = (sales_series > 0) & (sales_series.index == "coffee")

sales_series.loc[mask]
```

## LOGICAL OPERATORS AND METHODS

| Description | Python Operator | Pandas Method |
| :---: | :---: | :---: |
| Equal | == | .eq() |
| Not Equal | != | .ne() |
| Less Than or Equal | <= | .le() |
| Less Than | < | .lt() |
| Greater Than or Equal | >= | .ge() |
| Greater Than | > | .gt() |
| Membership Test | in | .isin() |
| Inverse Membership Test | not in | ~.isin() |

```python
sales_series.eq(5)
```

```python
sales_series.index.isin(["coffee", "tea"])
```

```python
~sales_series.index.isin(["coffee", "tea"])
```
