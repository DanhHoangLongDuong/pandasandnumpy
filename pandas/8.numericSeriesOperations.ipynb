## ARITHMETIC OPERATORS AND METHODS

| Description | Python Operator | Pandas Method |
| :---: | :---: | :---: |
| Addition | + | .add() |
| Subtraction | - | .sub(), .subtract() |
| Multiplication | * | .mul(), .multiply() |
| Division | / | .div(), .truediv(), .divide() |
| Floor Division | // | .floordiv() |
| Modulo | % | .mod() |
| Exponentiation | ** | .pow() |

```python
import pandas as pd 
import numpy as np
```

```python
my_series = pd.Series([1, np.nan, 2, 3, 4], index=["day0", "day1", "day2", "day3", "day4"])
my_series
```

```python
my_series + 1
```

```python
my_series.add(1, fill_value=0) #fill 0 into values that missing (NaN)
```

```python
my_series2 = my_series + 1
my_series2
```

```python
(my_series + my_series2) / 2
```
