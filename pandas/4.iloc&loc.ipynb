```python
#df.iloc[row position, column position] for dataframe
# .iloc[] is the preferred way to access values by their positional index
import pandas as pd
import numpy as np
```

```python
my_series = pd.Series(
    [0,1,2,3,4], index=["day0", "day1", "day2", "day3", "day4"]
)

my_series
```

```python
my_series.iloc[[1, 3, 4]]
```

```python
my_series.iloc[2]
```

```python
# .loc[] method is the preferred way to access values by their custom labels
# df.loc[row label, column label]
```

```python
sales_series = pd.Series(
    [0, 5, 155, 0, 518], index=["coffee", "bananas", "tea", "coconut", "sugar"], name="Sales"
)

sales_series
```

```python
sales_series.loc["tea"]
```

```python
sales_series.loc["coffee":"coconut"] #slices are inclusive when using custom labels, that's why coconut value also show up
```

```python
#drop index back to normal index
sales_series = sales_series.reset_index(drop=True)
sales_series
```

```python
sales_series.loc[:3]
```
