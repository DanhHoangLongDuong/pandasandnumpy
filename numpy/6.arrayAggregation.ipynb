```python
import numpy as np

rng = np.random.default_rng(12345)

randomArray = (rng.random(10) * 10).reshape(2, 5)
randomArray
```

```python
randomArray.sum(axis=0)
```

```python
randomArray.sum()
```

```python
print(np.median(randomArray)) #return median of array
print(np.percentile(randomArray, 80)) #return a value in the n^th percentile in an array
print(np.unique(randomArray)) #return the unique values in an array
print(np.sqrt(randomArray)) #return the square root of each value in an array
```
