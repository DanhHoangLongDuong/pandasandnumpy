```python
import numpy as np

rng = np.random.default_rng(12345)

sales_array = rng.integers(0, 100, size=(2, 5))
sales_array
```

```python
sales_array.sort() # sort row by row (axis = 1 by default)
sales_array
```

```python
sales_array.sort(axis=0) #sort by columns
sales_array
```

```python
#EXCERCISE
prices = np.array([5.99, 6.99, 22.49, 99.99, 4.99, 49.99])
price_tiers = np.array(["budget", "budget", "mid-tier", "luxury", "mid-tier", "luxury"])

prices.sort()
top3 = prices[-3::]
top3
```

```python
print(f"mean: {top3.mean()}")
print(f"min: {top3.min()}")
print(f"max: {top3.max()}")
print(f"median: {np.median(top3)}")
```
